## Data Preprocessing

```python
df = pd.read_csv('austinHousingData.csv')
df_filtered = df[['livingAreaSqFt', 'numOfBathrooms', 'avgSchoolRating', 'numOfBedrooms', 'numOfHighSchools', 'MedianStudentsPerTeacher', 'latestPrice']]

z = np.abs(stats.zscore(df_filtered))
threshold = 2.5
df_filtered_o = df_filtered[(z < threshold).all(axis=1)]
```

```python
X = df.iloc[:,:-1].to_numpy()
y = df.iloc[:,-1]

scaler = MinMaxScaler()
X = scaler.fit_transform(X)

oHe = OneHotEncoder()
y = df.iloc[:,-1].to_numpy().reshape([-1,1])
y = oHe.fit_transform(y, y=None).toarray()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.2)
```

## Model Fitting Using Keras

```python
model = tf.keras.models.Sequential()
#model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(20,activation='sigmoid'))
model.add(tf.keras.layers.Dense(16,activation='sigmoid'))
model.add(tf.keras.layers.Dense(40,activation='sigmoid'))
model.add(tf.keras.layers.Dense(32,activation='sigmoid'))
model.add(tf.keras.layers.Dense(17,activation='sigmoid'))

model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy','mse'])
model.optimizer.lr = 0.3
model.fit(X_train, y_train, epochs=3000)
```
